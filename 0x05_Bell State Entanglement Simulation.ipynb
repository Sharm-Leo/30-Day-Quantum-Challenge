📝 Project Overview

In this project, you will simulate a "Bell state"—the simplest example of quantum entanglement. When two qubits become entangled, their states are permanently linked. Measuring one qubit instantly determines the state of the other, no matter how far apart they are. You will build the circuit that creates this correlation and explore the foundational concepts behind the EPR paradox, quantum communication, and teleportation.


Task 0: The Multi-Qubit Environment

Set up your workspace for multiple interacting qubits.

[ ] Open a new Jupyter Notebook and import Qiskit.

[ ] Create a quantum circuit with 2 qubits and 2 classical bits.

[ ] Output: circuit.draw() shows two separate horizontal qubit wires (q0 and q1) and a classical register.

```python
import numpy as np

from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
```

```python
qc = QuantumCircuit(2,2)
qc.draw("mpl")
```

Task 1: Building the Bell State

Create the entanglement. The recipe for the most common Bell state (∣Φ+⟩) requires exactly two gates.

[ ] Apply a Hadamard gate (H-gate) to your first qubit (q0). This puts it in superposition.

[ ] Apply a Controlled-NOT gate (CNOT or CX-gate). Use q0 as the control qubit and q1 as the target qubit.

[ ] Output: Your circuit diagram visually shows the H-gate on q0, followed by the CNOT connection linking q0 to q1.

```python
# input state |00>

qc.h(0)
qc.cx(0,1)
qc.draw("mpl")
```

Task 2: Measuring the System

Extract the data from both entangled qubits.

[ ] Add measurement operations that map the outcome of q0 to the first classical bit, and q1 to the second classical bit.

[ ] Output: The circuit diagram displays measurement gauges on both qubit wires mapping down to the classical register.

```python
qc.measure(0,0)
qc.measure(1,1)
qc.draw("mpl")
```

Task 3: Simulating Spooky Action

Run the experiment to observe the quantum correlations.

[ ] Set up the QasmSimulator and execute the circuit for 1,000 shots.

[ ] Extract the result counts and plot a histogram using plot_histogram.

[ ] Output: Your histogram should reveal something incredible: you will see results for "00" (both qubits measured 0) and "11" (both qubits measured 1) roughly 50% of the time each.

[ ] Output (Analysis): Write a markdown cell explaining why the results "01" and "10" have a 0% probability. This proves that measuring the first qubit instantly dictated the state of the second.


```python
sim = AerSimulator(method='automatic')
result = sim.run(qc,shots=1000).result()
counts = result.get_counts(qc)
plot_histogram([counts], title='Simulating Entanglement', legend=['automatic'])
```
