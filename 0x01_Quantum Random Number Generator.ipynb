<h2>Task 1: Initialize the Quantum Circuit</h2>

Create the foundation for your quantum operations.

[ ] Define a quantum circuit with exactly 1 quantum bit (qubit) and 1 classical bit.

[ ] Output: Your code runs without errors, and you can visualize the empty circuit using circuit.draw().

```python
from qiskit import QuantumCircuit
from qiskit.circuit import QuantumRegister, ClassicalRegister
```

```python
qr = QuantumRegister(1)
cr = ClassicalRegister(1)
```

```python
combined_circ = QuantumCircuit(
    qr,cr
) 
```

```python
combined_circ.draw("mpl")
```

Run the circuit to generate your random numbers.

[ ] Set up the QasmSimulator (or a real IBM quantum backend).

[ ] Execute the circuit for multiple runs (e.g., shots=1000).

[ ] Output: Plot a histogram of the results. You should see a roughly 50% split between the '0' and '1' outcomes.

[ ] Output (Final): Write a small Python function that wraps this circuit and returns a single, true random '0' or '1' every time it is called.

```python
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
```

```python
def generate_n_random_number():
    shots = 1000
    sim_automatic = AerSimulator(method='automatic')
    job_automatic = sim_automatic.run(combined_circ, shots=shots)
    counts = job_automatic.result().get_counts(combined_circ)
    #plot_histogram([counts], title='Counts for automatic simulation output',legend=[ 'automatic'])

    random_bit = list(counts.keys())[0]
    return random_bit
```
